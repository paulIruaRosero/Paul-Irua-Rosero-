```python
#simulación del primer punto del parcial uno de señales
#comprobar con sympy las señales del punto (a)
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym

# Definir variables simbólicas
t, A, B, T = sym.symbols('t A B T', real=True)

# Definir señales
x1 = A * sym.exp(1j * sym.pi * t / T)
x2 = B * sym.exp(1j * 5 * sym.pi * t / T)

# Definir valores específicos para A, B y T
A_val = 2
B_val = 3
T_val = 10

# Convertir las expresiones simbólicas en funciones que se pueden evaluar
x1_func = sym.lambdify(t, x1.subs({A: A_val, T: T_val}), modules='numpy')
x2_func = sym.lambdify(t, x2.subs({B: B_val, T: T_val}), modules='numpy')

# Crear un arreglo de tiempo para evaluar las funciones
time = np.linspace(0, T_val, 1000)

# Evaluar las funciones en el arreglo de tiempo
x1_values = x1_func(time)
x2_values = x2_func(time)

# Graficar las señales en una misma figura
plt.figure(figsize=(10, 5))

plt.plot(time, np.real(x1_values), label='Re[x1]', color='blue')
plt.plot(time, np.imag(x1_values), label='Im[x1]', linestyle='--', color='blue')

plt.plot(time, np.real(x2_values), label='Re[x2]', color='red')
plt.plot(time, np.imag(x2_values), label='Im[x2]', linestyle='--', color='red')

plt.title('Señales x1 y x2')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# vamos a comprobar con la función simpy las integrales:
import sympy as sym

# Definir símbolos y constantes
t, A, B, T = sym.symbols('t A B T', real=True)
j = sym.I  # Definir la unidad imaginaria
W0 = 2 * sym.pi / T  # Frecuencia angular

# Definir las funciones a integrar
f1 = (A * sym.exp(j * W0 * t))**2
f2 = A * sym.exp(j * W0 * t) * B * sym.exp(j * 5 * W0 * t)
f3 = (B * sym.exp(j * 5 * W0 * t))**2

# Calculando  las integrales
integral1 = sym.integrate(f1, (t, 0, T)) / T
integral2 = -2 * sym.integrate(f2, (t, 0, T)) / T
integral3 = sym.integrate(f3, (t, 0, T)) / T

# Imprimir resultados
print("El valor de la integral 1 es:", integral1.simplify())
print("El valor de la integral 2 es:", integral2.simplify())
print("El valor de la integral 3 es:", integral3.simplify())
```

```python
# simulación del segundo punto del parcial 1
import numpy as np
import matplotlib.pyplot as plt

#Datos señal continua x1(t)
A1=3 #amplitud
f1=500 #frecuencia
T1=1/f1 #periodo

#Datos señal continua x2(t)
A2=5 #amplitud
f2=1000  #frecuencia
T2=1/f2 #periodo

#Datos señal continua x3(t)
A3=10 #amplitud
f3=5500 #frecuencia
T3=1/f3#periodo

#Periodo cuasiperiodica
T=1/5500

#Discretizar tiempo
ti=0 #tiempo inicial
tf=2*T #tiempo final
Fs=20*f3 #frecuencia de muestreo
Ts=1/Fs #tiempo de muestreo

#Crear vector discretizado
tv=np.arange(ti,tf,Ts)
print(tv)
print(tv.shape)

#Discretizar señal
xv=A1*np.cos((2*np.pi*f1*tv))+A2*np.sin((2*np.pi*f2*tv))+A3*np.cos((2*np.pi*f3*tv))
print(xv)
print(xv.shape)

#imprimir señal discretizada
plt.stem(tv,xv)
plt.plot(tv,xv)
plt.title("señal discretizada")
plt.xlabel("tiempo [seg]")
plt.ylabel("amplitud [V]")
plt.grid()
plt.show()
```

```python
# simulación del tercer punto del parcial 1
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

#Datos señal continua x1(t)
A1=20 #amplitud
f1=1/(6*np.pi) #frecuencia
T1=1/f1 #periodo

#Datos señal continua x2(t)
A2=20 #amplitud
f2=1/(8*np.pi) #frecuencia
T2=1/f2 #periodo

#Periodo cuasiperiodica
T=24*np.pi # se obtuvo hallando el mcm de T1 Y T2

#Discretizar tiempo
ti=0 #tiempo inicial
tf=2*T #tiempo final
Fs=20*f1 #frecuencia de muestreo cumpliendo Nyquist
Ts=1/Fs #tiempo de muestreo

#Crear vector discretizado
tv=np.arange(ti,tf,Ts)
print(tv)
print(tv.shape)

#Discretizar señal
xv=A1*np.cos((2*np.pi*f1*tv))+A2*np.sin((2*np.pi*f2*tv))
print(xv)
print(xv.shape)

#Cuantización de la señal

#Acondicionar la señal (modificar amplitud)
y_max=0.02
y_min=0.004
def modificar_amplitud (x,ymin,ymax):
  #Vamos a hallar la pendiente
  xmin=min(x) #Esta función me saca el minimo de la señal discretizada
  xmax=max(x) #Esta función me saca el maximo de la señal discretizada
  m=(ymax-ymin)/(xmax-xmin) #Aquí modificamos la pendiente para escalar la señal y que el dispositivo que trabaja de 4mA a 20mA la reciba de 4mA a 20mA
  c=ymax-m*xmax #Hallamos la c, porque la ecuación y=mx+c lo requiere y para que se mantenga la forma de la señal original
  yv=m*x+c #aqui calculamos la señal discretizada de 4mA a 20mA
  return yv

#vamos a pasarle los datos a la función modificar_amplitud para que nos calcule la señal acondicionada (yv)
yv= modificar_amplitud(xv,y_min,y_max)

#número de bits y vector de cuantización
nbits = 4
rmin = 0.004
rmax = 0.02
ve = np.linspace(rmin,rmax,2**nbits) # (rmax-rmin)/(2^#bits - 1) -> si se quiere incluir rmin y rmax

#distancias y cero pendiente a vector cuantizacion
d = cdist(yv.reshape(-1,1),ve.reshape(-1,1)) #se utiliza reshape(-1,1) para generar vectores columna como lo pide cdist
d.shape
#función cuantización punto a punto

def my_cuantizador(yn, vq) : #yn punto a #cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamaño vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1))#distancia yn a vector estados, reshape(-1,1) asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posición distancia min
  return vq[ind]


#recorrer vector cero y spam discretizado para cuantizar

for i in range(yv.shape[0]):
   if i==0: #primera muestra
     yq = my_cuantizador(yv[i],ve)
   else:
     yq = np.append(yq,my_cuantizador(yv[i],ve)) #apilar datos cuantizando si i>0


#graficar señal dsicreta


plt.stem(tv,yv, label='ycs_dis')#cs discreta

#estados de cuantización

for v in ve:
  plt.plot(tv,v*np.ones(len(tv)))

# graficar señal Digital (discreta + cuantizada)

plt.plot(tv,yq,c='y',marker='o',label='ydig', linewidth=2)

plt.legend()
plt.grid()
plt.xlabel('t[Seg]')
plt.ylabel('Amplitud [A]')
plt.show()
```
